# Analisis Lanjutan

```python
# --- Analisis RFM ---
import pandas as pd

# Gunakan data harian agar lebih mudah
rfm = df_day.groupby('registered').agg({
'instant': 'count', # jumlah hari pengguna aktif meminjam
'count': 'sum', # total peminjaman
'dteday': 'max' # tanggal terakhir aktivitas
}).reset_index()

# Ganti nama kolom agar sesuai dengan analisis RFM
rfm.columns = ['user_type', 'frequency', 'monetary', 'last_activity']

# Hitung recency (berapa hari sejak aktivitas terakhir)
latest_date = df_day['dteday'].max()
rfm['recency'] = pd.to_datetime(latest_date) - pd.to_datetime(rfm['last_activity'])
rfm['recency'] = rfm['recency'].dt.days

# Normalisasi skor RFM (semakin kecil recency = semakin bagus)
rfm['R_score'] = pd.qcut(rfm['recency'], 3, labels=[3,2,1])
rfm['F_score'] = pd.qcut(rfm['frequency'].rank(method='first'), 3, labels=[1,2,3])
rfm['M_score'] = pd.qcut(rfm['monetary'].rank(method='first'), 3, labels=[1,2,3])

# Total skor RFM
rfm['RFM_Score'] = rfm[['R_score', 'F_score', 'M_score']].sum(axis=1)
rfm
```

Insight-RFM Analysis

Analisis RFM dilakukan untuk memahami perilaku pengguna berdasarkan Recency (R), Frequency (F), dan Monetary (M) agar dapat mengidentifikasi tingkat loyalitas serta aktivitas pelanggan sepeda.

1. Recency (R)

• Nilai recency menunjukkan berapa hari sejak terakhir kali pengguna melakukan peminjaman.

• Semakin rendah nilai recency, semakin aktif pengguna.

• Berdasarkan data, sebagian besar pengguna aktif memiliki recency di bawah 100 hari, menandakan banyak pengguna yang masih rutin menggunakan layanan sepeda hingga mendekati akhir periode data (Oktober 2012).

• Namun terdapat juga kelompok pengguna dengan recency tinggi (>600 hari), yang menunjukkan mereka sudah tidak aktif lagi sejak awal tahun 2011.

2. Frequency (F)

• Menggambarkan seberapa sering pengguna meminjam sepeda dalam periode analisis.

• Nilai frequency tertinggi mencapai lebih dari 8.000 kali peminjaman, sementara nilai terendah hanya 20 kali.
Artinya terdapat kesenjangan yang signifikan antara pengguna kasual dan pengguna terdaftar.

• Sebagian besar pengguna memiliki frequency rendah–menengah (sekitar 1.000–4.000 kali), sedangkan pengguna dengan frequency tinggi kemungkinan besar pelanggan setia atau pengguna harian (misalnya untuk ke kantor atau kuliah).

3. Monetary (M)

• Monetary di sini direpresentasikan sebagai total jumlah peminjaman (karena dataset tidak memiliki nilai transaksi uang).

• Pengguna dengan skor monetary tertinggi berkontribusi besar terhadap total penggunaan layanan.

• Nilai monetary tertinggi mencapai 8.000+ peminjaman kumulatif, sementara sebagian besar pengguna memiliki nilai di bawah 2.000.

4. RFM Score

• Skor total RFM berkisar antara 3–9, dengan pengguna berskor tinggi (8–9) dianggap sebagai “Pelanggan Loyal” karena aktif, sering meminjam, dan memberikan kontribusi tinggi terhadap penggunaan layanan.

• Sebaliknya, pengguna dengan skor rendah (3–4) termasuk dalam kategori “At Risk” atau “Dormant Users”, yang jarang meminjam dan cenderung sudah tidak aktif.
